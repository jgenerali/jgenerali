```python
baseurl = "https://www.mudah.my/"

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}

adlist = []

url= "https://www.mudah.my/kuala-lumpur/cars-for-sale?o=1"
# There are total of 429 pages are there currently. Testing it out for first page
for x in range(1,2):
    k = requests.get('https://www.mudah.my/kuala-lumpur/cars-for-sale?o={}'.format(x)).text
    soup=BeautifulSoup(k,'html.parser')
    adlist = soup.find_all('div', attrs={"class" : "sc-eXEjpC bUbCth"})  #sc-jUpvKA lkaxYe or sc-eXEjpC bUbCth
    # Check the class code each time.
print(adlist)
```

```python
adlinks = []
adtitle = []
price = []
panel = []
condition = []
yearofm = []
verified = []
engine = []
milage = []
location = [] 
posteddate = []

data=[]
c=0
for ad in adlist:

    # Advertisement Link 
    try:
      adlink = ad.find("a").get('href') 
    except:
      adlink = None

    # Advertisement Title 
    try:
      adtitle = ad.find("a").get('title') 
    except:
      adtitle = None

    # CarPrice 
    try:
      price = ad.find("div",{"class":"sc-bsbRJL iDlyvB"}).text  #sc-bsbRJL iDlyvB or sc-jxGEyO jpjve
    except:
      price = None
    
    try:
      # CarCondition 
      condition = ad.find_all("div",{"class":"sc-esOvli zxCkq"})[0].text  #sc-esOvli zxCkq or 
      # Milage 
      milage = ad.find_all("div",{"class":"sc-esOvli zxCkq"})[1].text
      # MakingYear 
      yearofm = ad.find_all("div",{"class":"sc-esOvli zxCkq"})[2].text
      # Capacity   
      engine = ad.find_all("div",{"class":"sc-esOvli zxCkq"})[3].text 
    except:
      condition = None  
      milage = None
      yearofm = None  
      engine = None

    # Location    
    try:
      location = ad.find("span",{"class":"sc-gzOgki tpsvU"}).text  #sc-gzOgki tpsvU or 
    except:
      location = None

    # PostedDate    
    try:
      posteddate = ad.find("span",{"class":"sc-gxMtzJ enCjzM"}).text  #sc-gxMtzJ enCjzM or 
    except:
      posteddate = None      

    mudah = {"adlink":adlink,"adtitle":adtitle, "price":price, "condition":condition, "milage":milage, "yearofm":yearofm, "engine":engine, "location":location, "posteddate":posteddate}


    data.append(mudah)
    c=c+1
print("completed",c)
df = pd.DataFrame(data)
df.to_csv('mudah.csv',index=False)
!cp mudah.csv "drive/My Drive/"
```
